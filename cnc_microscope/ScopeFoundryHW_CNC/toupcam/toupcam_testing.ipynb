```python
from toupcam.camera import ToupCamCamera, get_number_cameras
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
print('num cams', get_number_cameras())
```

```python
cam = ToupCamCamera(resolution=5, cam_index=0)
#cam.do_awb()
cam.open()
```

```python
print(cam.bits)
print(cam.get_brightness())
print(cam.get_temperature_tint())
print(cam.get_exposure_time())
cam.set_temperature_tint(6503,500)
print(cam.get_temperature_tint())
print(cam.get_exposure_time())
```

```python
im = cam.get_pil_image()
im
```

```python
array = [np.array(item) for item in im.split()]
figure, ax = plt.subplots(1,3,figsize=(15,5))
for axis,channel in zip(ax,array):
    axis.imshow(channel,cmap='gray',clim=(0,256))
```
